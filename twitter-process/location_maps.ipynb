# Location Analysis

```python
import folium
from folium.plugins import HeatMap
import ijson
```

```python
# Create a dictionary to store tweet ID and location information
tweet_location_dict = {}
coordinates_list =[]

# Traverse the desired twitter.json file
with open("./output.json", 'r', encoding='utf-8') as f:
    
    # Create an ijson items iterator for the current process
    items = ijson.items(f, 'item',use_float=True)
    
    # Iterate over the items and process the tweets 
    for i, item in enumerate(items):
        sentiment = item["score"]
        longditude = item["centrex"]
        latitude = item["centrey"]
        coordinates_list.append((longditude,latitude,sentiment))
```

### Tweet Density

```python
from collections import Counter
# count number of tweets at each location
tweet_counts = Counter(coordinates_list)
tweet_counts
```

```python
# Create map centered on Australia
map = folium.Map(location=[-25, 135], zoom_start=4)

# Add markers for each location and score in the list
for tweet in coordinates_list:
    location = [tweet[1], tweet[0]]
    score = tweet[2]
    folium.Marker(location=location, popup=f"Score: {score}").add_to(map)

# Display the map
map
```
