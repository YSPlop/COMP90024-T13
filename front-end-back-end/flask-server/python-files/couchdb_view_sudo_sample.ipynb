```python
import json, couchdb
import geopandas as gpd
import pandas as pd

# Repuired to parse dictionary to MultiPolygon
from shapely.geometry import MultiPolygon, shape

# Connect to CouchDB-sudo
server = couchdb.Server('http://admin:password@172.26.131.88:5984')
db = server['sudo']

# Define view path
view_path_1 = '_design/sudo/_view/children_mother_education'
view_path_2 = '_design/sudo/_view/family_stress'
view_path_3 = '_design/sudo/_view/family_violence'
view_path_4 = '_design/sudo/_view/family_violence_patient'
view_path_5 = '_design/sudo/_view/housing_stress'
```

```python
# children_mother_education
data_1 = db.view(view_path_1, include_docs = True, reduce = False)
list_1 = []
for item in data_1:
    geo_str = str(item['doc']['geometry'])
    geo_json = json.loads(geo_str.replace("'", '"'))
    item['doc']['geometry'] = MultiPolygon(shape(geo_json))
    list_1.append(item['doc'])

df_1 = gpd.GeoDataFrame(list_1)
df_1.head()
```
